```python
from sklearn import datasets
import json
from openTSNE import TSNE
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
x, y = iris["data"], iris["target"]

embedding = TSNE().fit(x)
```

```python
data = pd.DataFrame(embedding, columns=["Comp1","Comp2"])
data["Label"] = y
```

```python
# TODO add the Target Information into  your json file
with open('tsne.json', 'w') as f:
    json.dump(np.array(data).tolist(), f)
```

```python
sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

sns.lmplot(x='Comp1',
           y='Comp2',
           data=data,
           fit_reg=False,
           legend=True,
           #size=10,
           hue='Label',
           scatter_kws={"s":200, "alpha":0.3}
          )

plt.title("TSNE", weight='bold').set_fontsize('14')
plt.xlabel('Comp 1')
plt.ylabel('Comp 2')
```

```python
from sklearn.manifold import SpectralEmbedding
model = SpectralEmbedding(n_components = 2, n_neighbors = 10)
result = model.fit_transform(x)
plt.figure(figsize=(20,15))
plt.scatter(result[:, 0], result[:, 1], c = y.astype(int), cmap = 'tab10', s = 50)
plt.title('UMAP', fontsize = 20)
plt.xlabel("Elem1", fontsize = 20)
plt.ylabel("Elem2", fontsize = 20)
plt.show();
```
